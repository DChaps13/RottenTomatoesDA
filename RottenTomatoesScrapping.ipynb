```python
import urllib.request as urllib2
from bs4 import BeautifulSoup
url = "https://www.rottentomatoes.com/franchise/marvel_cinematic_universe"
r = urllib2.urlopen(url)
html = r.read()
htmltext= html.decode('utf-8')
xkcdBsObj=BeautifulSoup(html,'lxml')
print(xkcdBsObj)
#print(type(htmltext))
```

```python
import re
htmltext2= htmltext.replace("\n","")
htmltext2= htmltext2.replace(" ","")
#genero= re.findall('<thscope="row">Gender</th><td>(.*?)</td>',infoStarkText, flags=re.S)[0]
stripsUrlM=re.findall('<strong><ahref="/m/(.*?)">',htmltext2)
stripsUrlTV=re.findall('<strong><ahref="/tv/(.*?)">',htmltext2)
```

```python
listUrl=[]
for strip in stripsUrlM:
    strip2M="https://www.rottentomatoes.com/m/"+strip
    listUrl.append(strip2M)
for strip in stripsUrlTV:
    strip2TV="https://www.rottentomatoes.com/tv/"+strip
    listUrl.append(strip2TV)
for url in listUrl:
    print(url)
```

```python
tomatoMeter=[]
namesStrip=[]
registers=[]
for url in listUrl:
    r2 = urllib2.urlopen(url)
    html3 = r2.read()
    html3text= html3.decode('utf-8')
    names=re.findall('data-type="title">(.*?)<',html3text,flags=re.S)
    name= names[0].strip()
    namesStrip.append(name)
    
    html3text= html3text.replace("\n","")
    html3text= html3text.replace(" ","")
    xkcdBsObj2=BeautifulSoup(html3,'lxml')
    #print(xkcdBsObj2)
    #encontraremos cada par de TomatoMeter [All Critics, TopCritics]
    strips3=re.findall('<spanclass="meter-valuesuperPageFontColor"><span>(.*?)</span>%</span>',html3text, flags= re.S)
    tomatoMeter.append(strips3)

registers.append(namesStrip)#columna de nombres
registers.append(tomatoMeter)#columna de tomato
#print(namesStrip)
#print(tomatoMeter)

```
